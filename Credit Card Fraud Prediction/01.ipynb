```python
import pandas as pd 
from sklearn.metrics import accuracy_score,classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('credit_card_fraud_dataset.csv')
```

```python
df=df.drop(columns='TransactionDate')
```

```python
df.head()
```

```python
lb=LabelEncoder()
df['TransactionType']=lb.fit_transform(df['TransactionType'])
df['Location']=lb.fit_transform(df['Location'])
```

```python
df['IsFraud'].value_counts()
```

```python
X=df.drop(columns='IsFraud',axis=1)
y=df['IsFraud']
```

```python
X_train,X_test,y_train,y_train=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
std=StandardScaler()
X_trian=std.fit_transform(X_train)
X_test=std.fit_transform(X_test)
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_trian,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
print(f'The accuracy score is : {accuracy_score(y_train,y_pred)}')
print(f'The classification report is : \n{classification_report(y_train,y_pred)}')
```
